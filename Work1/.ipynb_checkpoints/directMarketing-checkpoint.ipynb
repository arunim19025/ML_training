```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/DirectMarketing.csv")
df
```

```python
df["Catalogs"].value_counts()
```

```python
print(df["Gender"].value_counts())
print("----")
print(df["Married"].value_counts())
print("----")
print(df["OwnHome"].value_counts())
```

# Relation between Marrital status and children

```python
sns.countplot(x = df["Children"], hue=df["Married"])
```

# Relation between Age and HomeStatus

```python
sns.countplot(x = df["OwnHome"], hue=df["Age"])
```

# Relation between Age and Gender

```python
sns.countplot(x = df["Age"], hue=df["Gender"])
```

# Relation between HomeStatus and Marrital Status

```python
sns.countplot(x = df["Married"], hue=df["OwnHome"])
```

Most of the population comprises of middle in both the genders, male and female. There are more old females living than male and similarly more young females living than male. In total, more females are there than males. Number of Married and single people are almost the same. Most of the people own homes than stay on rent. Both single and married people owned almost equal number of children in each number.

# Relation between AmountSpent and Salary of a given person given the gender and age distribution with Home status 

```python
grid = sns.FacetGrid(df, col = "Age", row = "Gender", hue = "OwnHome")
grid.map(sns.scatterplot, "Salary", "AmountSpent")

# orange : Rent, blue : Own
```

# Relation between salary and the background history of that person

```python
sns.displot(x = df["Salary"], kind = "kde", hue = df["History"])
```

# Relation between Marrital status and salary and amount spent

```python
sns.violinplot(x = df["Married"], y = df["Salary"])
```

```python
sns.violinplot(x = df["Married"], y = df["AmountSpent"])
```

# Adding an additional column of percentage amount spent and savings

```python
df["Savings"] = df["Salary"] - df["AmountSpent"]
df["Percentage Spent"] = (df["AmountSpent"]/df["Salary"])*100
df
```

# Relation between Percentage Spent and Salary of a given person given the gender and age distribution with Home status 

```python
grid = sns.FacetGrid(df, col = "Age", row = "Gender", hue = "OwnHome")
grid.map(sns.scatterplot, "Percentage Spent", "Salary")

# orange : Rent, blue : Own
```

# Relation between Savings and Age

```python
sns.displot(x = df["Savings"], kind="kde", hue = df["Age"])
```
