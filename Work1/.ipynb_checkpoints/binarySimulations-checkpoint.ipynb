```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("data/binary.csv")
df.head()
```

```python
print("Mean of gre : "+str(df.gre.mean()))
print("Mean of gpa : "+str(df.gpa.mean()))
print("---------------------")
print("Mediean of gre : "+str(df.gre.median()))
print("Median of gpa : "+str(df.gpa.median()))
```

```python
df["rank"].value_counts()
```

```python
admission = df.loc[df["admit"]==1]
admission["rank"].value_counts()
```

# Rank and admission comparison

```python
sns.countplot(x = df["rank"], hue = df["admit"])
```

Conclusion :  As seen from the graph, the rank you maintain in your class directly contributes towards your chances in getting admission. Most of the students getting admission belong to top 2 ranks (total 87). It also shows how much the admission board favours rank 1 w.r.t to percentage selected (more than 50% in case of it)

# Performance of students on the basis of their gre score and gpa 

```python
sns.jointplot(x = df["gre"], y = df["gpa"], kind="hist")
```

Conclusion: We could see the most of the students who maintained a higher gpa track (greater than 3.75) record also had gre score (greater than 600), more significantly ones with gpa 4.0 and gre score 800. This is also applicable to mid range students where the maximum frequency can be seen in the range of gpa 3.25-3.4 and gre score 550-600. It also shows that very significant portion of low gpa students did perform better than mid and high gpa students in their gre scores. 

# Density plot of gre and gpa scores in terms of rank in their respective classes

```python
print(sns.displot(x = df["gpa"], kind = "kde", hue=df["rank"], palette = "Set1"))
print(sns.displot(x = df["gre"], kind = "kde", hue=df["rank"], palette = "Set1"))
```

Conclusion: Most of the 1st rankers had their gaussian plot tilted towards higher gpa and gre score which should be pretty obvious logically. But it is interesting to note that more number of rank 2 students performed better than rank 1 in their gre scores. Similarly rank 3 students performed equally well in case of their gpa.

# Scatter plot depicting the chances of admission w.r.t gre score and gpa

```python
df.loc[(df["gre"] == 800) & (df["gpa"] == 4.00)]
```

```python
df.loc[(df["gre"] <= df.gre.mean()) & (df["gpa"] <= df.gpa.mean()) & (df["admit"] == 1)]["rank"].value_counts()
```

```python
sns.scatterplot(x = df["gpa"], y = df["gre"], hue=df["admit"], style = df["admit"])
```

Conclusion: In most cases, admission process was inclined towards consistent performers more than students who had higher gpa. GRE score matters a lot more than gpa as clear from the graph. If the gre score is less, this should be compensated with above average/median gpa score. In very rate situations, less gpa and gre score are considered but in that case in most cases they had their ranks higher in their respective class
